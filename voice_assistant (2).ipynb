Define Speak Function

```python
from gtts import gTTS
from IPython.display import Audio, display

def speak(text):
    """Convert text to speech and play it."""
    print("Assistant:", text)
    tts = gTTS(text)
    tts.save("response.mp3")
    display(Audio("response.mp3", autoplay=True))
```

Define AI Response Function (Offline)

```python
def ai_response(prompt):
    """Basic logic without API."""
    prompt = prompt.lower()

    if "your name" in prompt:
        return "I am your voice assistant."
    elif "time" in prompt:
        from datetime import datetime
        return f"The time is {datetime.now().strftime('%I:%M %p')}."
    elif "hello" in prompt or "hi" in prompt:
        return "Hello! How can I help you today?"
    elif "joke" in prompt:
        return "Why don't scientists trust atoms? Because they make up everything!"
    elif "bye" in prompt or "exit" in prompt:
        return "Goodbye!"
    else:
        return "Sorry, I don't know how to respond to that."
```

Start the Assistant Loop

```python
def ask_assistant():
    """Main chat loop."""
    print("Type your message (type 'exit' to stop):")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit", "bye"]:
            speak("Goodbye!")
            break
        response = ai_response(user_input)
        speak(response)

# Start assistant
ask_assistant()
```
